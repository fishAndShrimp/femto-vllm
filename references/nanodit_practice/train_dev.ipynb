```python
my_transform = transforms.Compose(
    [
        transforms.Resize((64, 64)),
        transforms.ToTensor(),
        transforms.Normalize([0.5], [0.5]),
    ]
)
train_dataset = torchvision.datasets.ImageFolder(
    "datasets/butterflies", transform=my_transform
)
```

```python
t_img = train_dataset[999][0]
print(t_img.shape)
transforms.ToPILImage()(t_img * 0.5 + 0.5)
```
